```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# creating a mock dataset
data = {
    'Date': ['2023-01-15', '2023-01-18', '2023-02-10', '2023-02-22', '2023-03-05', '2023-03-10', '2023-03-25'],
    'Product': ['Laptop', 'Mouse', 'Laptop', 'Keyboard', 'Mouse', 'Laptop', 'Keyboard'],
    'Price': [1000, 25, 1000, 50, 25, 1000, 50],
    'Quantity': [1, 10, 2, 5, 15, 1, 10],
    'Customer_Age': [25, 34, 25, 45, 34, 25, 45] # Added demographic data
}
```

```python
# convert dictionary to a Pandas as dataframe (like an Excel sheet)
df=pd.DataFrame(data)
```

```python
#Preview the Data
print("--- Raw Data ---")
print(df.head())
```

```python
#1. Convert Data column to datetime objects 
df['Date'] = pd.to_datetime(df['Date'])
    #2. Create a new column: 'Total_Sales' (Price * Quantity)
df['Total_Sales'] = df['Price'] * df['Quantity']
print ("\n--- Cleaned Data with Total Sales ---")
print(df)
```

```python
#Analysis part that Answering Questions 
# Which product generates the most revenue?
product_sales =df.groupby('Product')['Total_Sales'].sum()
print("\n--- Sales by Product ---")
print(product_sales)
```

```python
#what was the best month for sales?
#Exract month name 
df['month'] = df['Date'].dt.month_name()
monthly_sales = df.groupby('month')['Total_Sales'].sum()
print("\n---Sales by Month ---")
print(monthly_sales)
```

```python
#visualization
#create a figure
plt.figure(figsize=(10,6))

```

```python
#plot the data
#product_sales.index is the names (laptop, Mouse), values is the money
plt.bar(product_sales.index, product_sales.values, color='skyblue')
```

```python
#add labels 
plt.title('Total Revenue by Product')
plt.xlabel('Product name')
plt.ylabel('Revenue($)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
#show the plot
plt.show()
```

```python
# the condition: checks every row. Returns true if Age < 30, False otherwise.
under_30_mask =df['Customer_Age'] < 30
#Apply the mask to the dataframe 
young_customers_df =df[under_30_mask]
print("--- Sales to Customers Under 30 ---")
print(young_customers_df)
```

```python
young_customers_df = df[df['Customer_Age'] < 30]
```

```python
# Condition 1: Age < 30
# Condition 2: Product == 'Laptop'
target_sales = df[(df['Customer_Age'] < 30) & (df['Product'] == 'Laptop')]
print("\n--- Laptops sold to Under-30s ---")
print(target_sales)
```

```python
young_revenue = young_customers_df['Total_Sales'].sum()
print(f"\nTotal Revenue from Under_30s: ${young_revenue}")
```

```python
# Save the filtered data to a CSV file
# index=False means "don't write the row numbers (0,1,2) into the file"
young_customers_df.to_csv('young_customer_report.csv', index=False)
print("CSV saved successfully!")
```
